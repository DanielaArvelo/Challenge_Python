### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Convertimos las columnas "Precio" en listas
precios_tienda1 = tienda["Precio"].tolist()
precios_tienda2 = tienda2["Precio"].tolist()
precios_tienda3 = tienda3["Precio"].tolist()
precios_tienda4 = tienda4["Precio"].tolist()

# Calculamos la suma (ingreso total) de cada tienda
ingreso_tienda1 = sum(precios_tienda1)
ingreso_tienda2 = sum(precios_tienda2)
ingreso_tienda3 = sum(precios_tienda3)
ingreso_tienda4 = sum(precios_tienda4)

# Mostramos el resultado
print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)
```

```python
import matplotlib.pyplot as plt

ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"])
plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso Total")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

# 2. Ventas por categoría

```python
# Obtenemos la lista de categorías para la Tienda 1
categorias_tienda1 = tienda["Categoría del Producto"].tolist()

# Creamos un diccionario para contar las ventas por categoría
conteo_categorias_tienda1 = {}

for categoria in categorias_tienda1:
    if categoria in conteo_categorias_tienda1:
        conteo_categorias_tienda1[categoria] += 1
    else:
        conteo_categorias_tienda1[categoria] = 1

# Mostramos el resultado
print("Ventas por categoría en Tienda 1:")
for categoria, cantidad in conteo_categorias_tienda1.items():
    print(f"{categoria}: {cantidad}")
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras horizontales para ventas por categoría en Tienda 1
plt.figure(figsize=(10, 6))
plt.barh(list(conteo_categorias_tienda1.keys()), list(conteo_categorias_tienda1.values()), color="#1f77b4")
plt.title("Ventas por Categoría - Tienda 1")
plt.xlabel("Cantidad de Ventas")
plt.ylabel("Categoría")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Tienda 1
calificaciones_tienda1 = tienda["Calificación"].tolist()
suma_calificaciones_tienda1 = sum(calificaciones_tienda1)
cantidad_calificaciones_tienda1 = len(calificaciones_tienda1)
media_calificacion_tienda1 = suma_calificaciones_tienda1 / cantidad_calificaciones_tienda1 if cantidad_calificaciones_tienda1 > 0 else 0
print("Calificación media Tienda 1:", media_calificacion_tienda1)

# Tienda 2
calificaciones_tienda2 = tienda2["Calificación"].tolist()
suma_calificaciones_tienda2 = sum(calificaciones_tienda2)
cantidad_calificaciones_tienda2 = len(calificaciones_tienda2)
media_calificacion_tienda2 = suma_calificaciones_tienda2 / cantidad_calificaciones_tienda2 if cantidad_calificaciones_tienda2 > 0 else 0
print("Calificación media Tienda 2:", media_calificacion_tienda2)

# Tienda 3
calificaciones_tienda3 = tienda3["Calificación"].tolist()
suma_calificaciones_tienda3 = sum(calificaciones_tienda3)
cantidad_calificaciones_tienda3 = len(calificaciones_tienda3)
media_calificacion_tienda3 = suma_calificaciones_tienda3 / cantidad_calificaciones_tienda3 if cantidad_calificaciones_tienda3 > 0 else 0
print("Calificación media Tienda 3:", media_calificacion_tienda3)

# Tienda 4
calificaciones_tienda4 = tienda4["Calificación"].tolist()
suma_calificaciones_tienda4 = sum(calificaciones_tienda4)
cantidad_calificaciones_tienda4 = len(calificaciones_tienda4)
media_calificacion_tienda4 = suma_calificaciones_tienda4 / cantidad_calificaciones_tienda4 if cantidad_calificaciones_tienda4 > 0 else 0
print("Calificación media Tienda 4:", media_calificacion_tienda4)
```

```python
import matplotlib.pyplot as plt

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones_medias = [3.9767, 4.0373, 4.0483, 3.9958]

plt.figure(figsize=(8, 5))
plt.scatter(tiendas, calificaciones_medias, color="#1f77b4", s=150)
plt.title("Calificación Promedio por Tienda (Zoom)")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.ylim(3.9, 4.1)  # Zoom para resaltar diferencias
plt.grid(True, linestyle="--", alpha=0.7)

# Mostrar etiquetas de valor encima de cada punto
for i, val in enumerate(calificaciones_medias):
    plt.text(i, val + 0.005, f"{val:.3f}", ha='center', va='bottom')

plt.show()
```

# 4. Productos más y menos vendidos

```python
# Lista de productos de la tienda 1
productos_tienda1 = tienda["Producto"].tolist()

# Contamos las ventas de cada producto
conteo_productos_tienda1 = {}

for producto in productos_tienda1:
    if producto in conteo_productos_tienda1:
        conteo_productos_tienda1[producto] += 1
    else:
        conteo_productos_tienda1[producto] = 1

# Convertimos el diccionario en una lista de tuplas (producto, cantidad)
productos_ordenados1 = sorted(conteo_productos_tienda1.items(), key=lambda x: x[1], reverse=True)

# Top 3 más vendidos
print("Tienda 1 - Top 3 productos más vendidos:")
for producto, cantidad in productos_ordenados1[:3]:
    print(f"{producto}: {cantidad} ventas")

# Top 3 menos vendidos
print("Tienda 1 - Top 3 productos menos vendidos:")
for producto, cantidad in productos_ordenados1[-3:]:
    print(f"{producto}: {cantidad} ventas")
```

```python
import matplotlib.pyplot as plt

# Extraemos los top 3 más vendidos y top 3 menos vendidos
top_mas_vendidos = productos_ordenados1[:3]
top_menos_vendidos = productos_ordenados1[-3:]

# Productos y cantidades para cada grupo
productos_mas = [p[0] for p in top_mas_vendidos]
cantidades_mas = [p[1] for p in top_mas_vendidos]

productos_menos = [p[0] for p in top_menos_vendidos]
cantidades_menos = [p[1] for p in top_menos_vendidos]

# Configuración del gráfico
plt.figure(figsize=(12, 6))

# Barras para más vendidos
plt.bar(productos_mas, cantidades_mas, color='green', label='Top 3 Más Vendidos')

# Barras para menos vendidos
plt.bar(productos_menos, cantidades_menos, color='red', label='Top 3 Menos Vendidos')

plt.title('Productos Más y Menos Vendidos - Tienda 1')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 5. Envío promedio por tienda

```python
# Lista de costos de envío
envios_tienda1 = tienda["Costo de envío"].tolist()

# Suma total de envíos
suma_envios_tienda1 = sum(envios_tienda1)

# Número de envíos
cantidad_envios_tienda1 = len(envios_tienda1)

# Cálculo del promedio
if cantidad_envios_tienda1 > 0:
    promedio_envio_tienda1 = suma_envios_tienda1 / cantidad_envios_tienda1
else:
    promedio_envio_tienda1 = 0

print("Tienda 1 - Costo de envío promedio:", promedio_envio_tienda1)
```

```python
# Tienda 2
envios_tienda2 = tienda2["Costo de envío"].tolist()
suma_envios_tienda2 = sum(envios_tienda2)
cantidad_envios_tienda2 = len(envios_tienda2)
promedio_envio_tienda2 = suma_envios_tienda2 / cantidad_envios_tienda2 if cantidad_envios_tienda2 > 0 else 0
print("Tienda 2 - Costo de envío promedio:", promedio_envio_tienda2)
```

```python
# Tienda 3
envios_tienda3 = tienda3["Costo de envío"].tolist()
suma_envios_tienda3 = sum(envios_tienda3)
cantidad_envios_tienda3 = len(envios_tienda3)
promedio_envio_tienda3 = suma_envios_tienda3 / cantidad_envios_tienda3 if cantidad_envios_tienda3 > 0 else 0
print("Tienda 3 - Costo de envío promedio:", promedio_envio_tienda3)
```

```python
# Tienda 4
envios_tienda4 = tienda4["Costo de envío"].tolist()
suma_envios_tienda4 = sum(envios_tienda4)
cantidad_envios_tienda4 = len(envios_tienda4)
promedio_envio_tienda4 = suma_envios_tienda4 / cantidad_envios_tienda4 if cantidad_envios_tienda4 > 0 else 0
print("Tienda 4 - Costo de envío promedio:", promedio_envio_tienda4)
```
